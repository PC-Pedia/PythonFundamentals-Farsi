## **آموزش جنگو**

🧠 Django یک چارچوب توسعه وب (Web framework) بر پایه‌ی زبان برنامه‌نویسی پایتون (Python) است. این چارچوب به منظور تسهیل و تسریع فرآیند توسعه وب‌سایت‌ها و برنامه‌های تحت وب طراحی شده است. Django یکی از پرکاربردترین و قدرتمندترین چارچوب‌های وب در جهان محسوب می‌شود و از طراحی منظم، مستندات کامل، امکانات فراوان و جوامع فعال برخوردار است.

ویژگی‌های Django عبارتند از:

1. **ORM (Object-Relational Mapping):** Django از یک سیستم ORM برخوردار است که به توسعه‌دهندگان این امکان را می‌دهد که با استفاده از شیء‌ها و کلاس‌ها به جای SQL، با پایگاه داده تعامل داشته باشند.

2. **ساختار MVC (Model-View-Controller):** Django از مدل-نما-کنترلر به عنوان الگوی طراحی استفاده می‌کند. این الگو به توسعه‌دهندگان این امکان را می‌دهد تا بخش‌های مختلف برنامه را جداگانه مدیریت و توسعه دهند.

3. **سیستم اداره‌ی کاربران و دسترسی (Authentication and Authorization):** Django یک سیستم مدیریت کاربران را داراست که احراز هویت و کنترل دسترسی‌ها را فراهم می‌کند.

4. **پشتیبانی از فرم‌ها (Forms):** Django ابزارها و کلاس‌هایی برای ایجاد و پردازش فرم‌ها فراهم کرده است که توسعه‌دهندگان را از زحمت نوشتن کد HTML و پردازش دستی فرم‌ها خارج می‌کند.

5. **سیستم مدیریت پایگاه داده (Database Management):** Django از یک سیستم مدیریت پایگاه داده به نام "django ORM" برای اتصال به پایگاه داده و انجام عملیات CRUD استفاده می‌کند.

6. **سیستم URL ها و Routing:** Django یک سیستم قوی برای مدیریت URL ها و مسیریابی (routing) دارد که از آن برای مدیریت درخواست‌های وب به بخش‌های مختلف برنامه استفاده می‌شود.

7. **قابلیت‌های امنیتی:** Django از ویژگی‌های امنیتی همچون جلوگیری از حملات ایجکشن SQL (SQL injection) و حفاظت در برابر حملات Cross-Site Scripting (XSS) پشتیبانی می‌کند.

Django با ارائه‌ی این ویژگی‌ها و امکانات به توسعه‌دهندگان این امکان را می‌دهد که برنامه‌های تحت وب پویا، امن و مقیاس‌پذیر را با سرعت بسازند.

Django یک چارچوب توسعه وب است که از مدل-نما-کنترلر (Model-View-Controller) به عنوان الگوی طراحی استفاده می‌کند، اما به جای این الگو، از یک الگوی مشابه به نام مدل-نما-کنترلر-تمپلیت (Model-View-Template) یا MVT استفاده می‌کند. در Django، اجزای اصلی عبارتند از:

1. **مدل (Model):** مسئول مدیریت و انجام عملیات مربوط به داده‌ها و پایگاه داده است. مدل‌ها تعیین می‌کنند چگونه داده‌ها در پایگاه داده ذخیره می‌شوند و چگونه با آن‌ها برخورد شود.

2. **نما (View):** مسئول پردازش و ارتباط با کاربر است. نماها وظیفه نمایش داده‌ها و درخواست‌های کاربر را برعهده دارند. آن‌ها با استفاده از اطلاعاتی که از مدل می‌گیرند، پاسخ به درخواست‌ها و ارائه داده‌ها به کاربر را انجام می‌دهند.

3. **کنترلر (Controller):** در Django، کنترلر به عنوان بخشی از فرآیند نما تعبیه شده است و وظیفه مدیریت جریان کار و تعیین اطلاعات مربوط به داده‌ها است. این قسمت بیشتر در فرآیند نما یک نقش کمکی دارد.

4. **تمپلیت (Template):** تمپلیت‌ها مسئول نمایش داده‌ها به کاربر هستند. آن‌ها به عنوان یک لایه میانی بین مدل و نما عمل می‌کنند و اجازه می‌دهند داده‌ها را به صورت داینامیک در HTML و دیگر فرمت‌های خروجی نمایش داده شوند.

فرآیند کار Django به شرح زیر است:

1. **تعریف مدل‌ها:** در ابتدا، شما مدل‌ها را بر اساس داده‌هایی که می‌خواهید در پایگاه داده ذخیره کنید، تعریف می‌کنید.

2. **ایجاد URL ها:** شما URL هایی را تعریف می‌کنید که به کنترلرهای Django اشاره داشته باشند.

3. **تعریف نماها و کنترلرها:** شما نماها و کنترلرهای خود را تعریف می‌کنید که به کمک مدل‌ها داده‌ها را دریافت کرده، پردازش می‌کنند و نتیجه را به کاربر ارائه می‌دهند.

4. **تعریف تمپلیت‌ها:** شما تمپلیت‌ها را برای نمایش داده‌ها به کاربر تعریف می‌کنید.

5. **اجتماع تمپلیت و نما:** نماها از تمپلیت‌ها برای نمایش داده‌ها به کاربر استفاده می‌کنند.

6. **مدیریت درخواست‌ها و پاسخ‌ها:** Django به عنوان یک واسط بین کاربر و پایگاه داده عمل کرده و اطلاعات را بین مدل، نما و تمپلیت مدیریت می‌کند.

این چهارچوب با ارائه‌ی ابزارها و پیش‌فرض‌های زیادی به توسعه‌دهندگان کمک می‌کند تا به سرعت و به شیوه‌ای کارآمد و ایمن برنامه‌های تحت وب را ایجاد کنند.
