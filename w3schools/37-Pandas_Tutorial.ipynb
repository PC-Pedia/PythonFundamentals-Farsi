Pandas - Cleaning Empty Cells

<section style="direction: rtl; text-align: justify; margin: 30px;">

## **پاکسازی سلول‌های خالی**

سلول‌های خالی به طور بالقوه می‌توانند هنگام تجزیه و تحلیل داده‌ها، نتیجه اشتباهی به شما بدهند.


### **حذف ردیف‌های خالی**

یکی از راه های مقابله با سلول های خالی حذف ردیف هایی است که حاوی سلول های خالی هستند.
این معمولاً مشکلی ندارد، زیرا مجموعه داده‌ها می‌توانند بسیار بزرگ باشند و حذف چند ردیف تأثیر زیادی در نتیجه نخواهد داشت.

</section>

```python
import pandas as pd

df = pd.read_csv('data.csv')

new_df = df.dropna()

print(new_df.to_string())
```

<section style="direction: rtl; text-align: justify; margin: 30px;">

توجه: به طور پیش‌فرض، متد dropna() یک DataFrame جدید برمی‌گرداند و نسخه اصلی را تغییر نخواهد داد.

اگر می خواهید DataFrame اصلی را تغییر دهید، از آرگومان inplace = True استفاده کنید:


</section>

```python
import pandas as pd

df = pd.read_csv('data.csv')

df.dropna(inplace=True)

print(df.to_string())
```

<section style="direction: rtl; text-align: justify; margin: 30px;">

توجه: اکنون dropna (inplace = True) DataFrame جدیدی را برنمی‌گرداند، اما تمام ردیف‌های حاوی مقادیر NULL را از DataFrame اصلی حذف می‌کند.

</section>

Replace Empty Values

<section style="direction: rtl; text-align: justify; margin: 30px;">

### **جایگذاری مقادیر خالی**

روش دیگر برای مقابله با سلول های خالی، درج یک مقدار جدید به جای آن است.
به این ترتیب شما مجبور نیستید تمام سطرها را فقط به دلیل چند سلول خالی حذف کنید.
متد fillna() به ما اجازه می دهد تا سلول های خالی را با یک مقدار جایگزین کنیم:

</section>

```python
import pandas as pd

df = pd.read_csv('data.csv')

df.fillna(130, inplace=True)
```

Replace Only For Specified Columns

<section style="direction: rtl; text-align: justify; margin: 30px;">

### **فقط برای ستون های مشخص شده جایگزین کنید**

مثال بالا جایگزین تمام سلول های خالی در کل Data Frame می شود.
فقط برای جایگزینی مقادیر خالی برای یک ستون، نام ستون را برای DataFrame مشخص کنید:

</section>

```python
import pandas as pd

df = pd.read_csv('data.csv')

df["Calories"].fillna(130, inplace=True)

# print(df.head())
```

Replace Using Mean, Median, or Mode

<section style="direction: rtl; text-align: justify; margin: 30px;">

### **با استفاده از Mean، Median یا Mode جایگزین کنید**

یک روش رایج برای جایگزینی سلول های خالی، محاسبه میانگین، میانه یا مقدار حالت ستون است.
پانداس از متدهای mean() median() و mode() برای محاسبه مقادیر مربوطه برای یک ستون مشخص استفاده می کنند:

</section>

```python
import pandas as pd

df = pd.read_csv('data.csv')

x = df["Calories"].mean()

df["Calories"].fillna(x, inplace=True)

# print(df.head())
```

<section style="direction: rtl; text-align: justify; margin: 30px;">

☝🏽 میانگین = مقدار متوسط ​​(مجموع همه مقادیر تقسیم بر تعداد مقادیر).

</section>

```python
import pandas as pd

df = pd.read_csv('data.csv')

x = df["Calories"].median()

df["Calories"].fillna(x, inplace=True)

# print(df.head())
```

<section style="direction: rtl; text-align: justify; margin: 30px;">

☝🏽 میانه = مقدار در وسط، بعد از اینکه همه مقادیر را به صورت صعودی مرتب کردید.

</section>

```python
import pandas as pd

df = pd.read_csv('data.csv')

x = df["Calories"].mode()[0]

df["Calories"].fillna(x, inplace=True)

print(df.head())
```

<section style="direction: rtl; text-align: justify; margin: 30px;">

☝🏽 Mode = مقداری که بیشتر ظاهر می شود.

</section>

## **پاکسازی داده‌های با فرمت اشتباه**

سلول هایی با داده هایی با فرمت اشتباه می توانند تجزیه و تحلیل داده ها را دشوار یا حتی غیرممکن کنند.

</section>

```python
import pandas as pd

df = pd.read_csv('data2.csv')

df['Date'] = pd.to_datetime(df['Date'])

print(df.to_string())
```

<section style="direction: rtl; text-align: justify; margin: 30px;">

نتیجه تبدیل در مثال بالا به ما یک مقدار NaT داد که می‌توان آن را به عنوان یک مقدار NULL استفاده کرد و می‌توانیم با استفاده از روش dropna() ردیف را حذف کنیم.

</section>

```python
df.dropna(subset=['Date'], inplace= True)
```
