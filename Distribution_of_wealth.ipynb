```python
import matplotlib.pyplot as plt
import random
```

```python
people = [100 for i in range(100)]
```

```python
for iteration in range(1000 - 1):
    p1, p2 = random.randint(0, 99), random.randint(0, 99)
    if p1 != p2:
        people[p1] += 1
        people[p2] -= 1
```

```python
plt.plot(people, marker='.',)
constant = [100 for i in range(100)]
plt.plot(constant, color='r')
```

```python
def people_wealth(n=10000):
    people = [100 for i in range(100)]

    for iteration in range(n - 1):
        p1, p2 = random.randint(0, 99), random.randint(0, 99)
        if p1 != p2:
            people[p1] += 1
            people[p2] -= 1

    plt.plot(people, marker='.',)
    constant = [100 for i in range(100)]
    plt.plot(constant, color='r')


people_wealth(n=100_000)
```
