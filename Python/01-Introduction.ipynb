### **پایتون چیست؟**
پایتون یک زبان برنامه‌نویسی محبوب است. این زبان توسط گوئیدو ون رسوم ایجاد شده و در سال 1991 منتشر شد.

از آن برای موارد زیر استفاده می‌شود:

- توسعه وب (سمت سرور)
- توسعه نرم‌افزار
- ریاضیات
- اسکریپت‌نویسی سیستم

### **پایتون چه کارهایی انجام می‌دهد؟**
- پایتون می‌تواند برای ایجاد برنامه‌های وب روی یک سرور مورد استفاده قرار گیرد.
- پایتون می‌تواند به همراه نرم‌افزار برای ایجاد جریان‌های کاری استفاده شود.
- پایتون قادر است به سیستم‌های پایگاه داده متصل شود. همچنین، فایل‌ها را می‌تواند خوانده و تغییر دهد.
- پایتون می‌تواند برای کنترل داده‌های بزرگ و انجام محاسبات پیچیده استفاده شود.
- پایتون می‌تواند برای پروتوتایپ‌سازی سریع یا توسعه نرم‌افزارهای آماده برای تولید استفاده شود.

### **چرا پایتون؟**
- پایتون بر روی پلتفرم‌های مختلف (ویندوز، مک، لینوکس، Raspberry Pi و غیره) کار می‌کند.
- پایتون دارای یک دستور زبان ساده است که شباهت‌هایی به زبان انگلیسی دارد.
- پایتون از دستورات جدید برای تکمیل یک دستور استفاده می‌کند، برخلاف زبان‌های برنامه‌نویسی دیگر که اغلب از نقطه‌ویرگول یا پرانتز استفاده می‌کنند.
- پایتون به وابستگی به فاصله‌گذاری و استفاده از فضای خالی برای تعریف دامنه استفاده می‌کند؛ به عنوان مثال دامنه حلقه‌ها، توابع و کلاس‌ها. زبان‌های برنامه‌نویسی دیگر اغلب از پرانتز‌های زبانه‌ای برای این منظور استفاده می‌کنند.

### **مواردی که باید بدانید**
- جدیدترین نسخه اصلی پایتون، پایتون 3 است که در این آموزش از آن استفاده خواهیم کرد. با این حال، پایتون 2، اگرچه به جز به‌روزرسانی‌های امنیتی، به‌روز نشده است اما همچنان محبوب است.
- در این آموزش، پایتون به صورت متنی در یک ویرایشگر متنی نوشته خواهد شد. امکان نوشتن پایتون در یک محیط توسعه یکپارچه مانند Thonny، PyCharm، NetBeans یا Eclipse نیز وجود دارد که بخصوص در مدیریت مجموعه‌های بزرگی از فایل‌های پایتون مفید است.

### **نحوه نوشتن پایتون در مقایسه با زبان‌های برنامه‌نویسی دیگر**
- پایتون برای خوانا بودن طراحی شده است و برخی شباهت‌ها به زبان انگلیسی با تأثیراتی از ریاضیات دارد.
- پایتون از خطوط جدید برای تکمیل یک دستور استفاده می‌کند، در مقابل زبان‌های برنامه‌نویسی دیگر که اغلب از نقطه‌ویرگول یا پرانتز استفاده می‌کنند.
- پایتون به فاصله‌گذاری و استفاده از فضای خالی برای تعریف دامنه اعتماد می‌کند؛ مانند دامنه حلقه‌ها، توابع و کلاس‌ها. زبان‌های برنامه‌نویسی دیگر اغلب از پرانتز‌های خمیده برای این منظور استفاده می‌کنند.
